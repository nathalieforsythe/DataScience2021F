### Set-up
Downloads libraries and data

```python
import pandas as pd
import seaborn as sb
from sklearn.model_selection import cross_val_score

titanic_data = pd.read_csv('train.csv', index_col = 'PassengerId')
test_data = pd.read_csv('test.csv', index_col = 'PassengerId')
titanic_data
```

<a id='data_cleaning'></a>
### Data cleaning

- Looking for null values to see which columns need to be altered and which can be used as they are
- titanic_data: Embarked has 2 missing values that will need to be imputed
- test_dat: Fare is has 1 null that will need to be imputed

```python
titanic_data.info()
test_data.info()
```

Replaces male/female with integers in both data sets so that Sex can be used to make predictions

```python
titanic_data['Sex'] = titanic_data['Sex'].replace(['female'], 0)
titanic_data['Sex'] = titanic_data['Sex'].replace(['male'], 1)
test_data['Sex'] = test_data['Sex'].replace(['female'], 0)
test_data['Sex'] = test_data['Sex'].replace(['male'], 1)
test_data
```

- Creates a bar graph of the number of passengers who embarked from each location to find distribution of places that passengers embarked from.
- Most people had embarked from S

```python
embarked_loc = titanic_data.groupby(['Embarked']).count()
embarked_loc = embarked_loc['Survived']
sb.barplot(x = embarked_loc.index, y = embarked_loc)
```

<a id='imputation'></a>
### Imputation

Replaces missing Embarked values with the mode

```python
# I couldn't get this to work by using mode() - everything came back null
# fillna: https://stackoverflow.com/questions/40619445/how-to-replace-na-values-with-mode-of-a-dataframe-column-in-python/40621330
titanic_data['Embarked'].fillna('S', inplace=True)
```

Replaces names for each place in the Embarkment column with a number so that it can be used in predictions.

```python
titanic_data['Embarked'] = titanic_data['Embarked'].replace(['C'], 0)
titanic_data['Embarked'] = titanic_data['Embarked'].replace(['Q'], 1)
titanic_data['Embarked'] = titanic_data['Embarked'].replace(['S'], 2)
test_data['Embarked'] = test_data['Embarked'].replace(['C'], 0)
test_data['Embarked'] = test_data['Embarked'].replace(['Q'], 1)
test_data['Embarked'] = test_data['Embarked'].replace(['S'], 2)
test_data
```

Replaces nulls in Age with medians 

```python
age_med = titanic_data['Age'].median() # 28.0
test_age_med = test_data['Age'].median() # 27.0
titanic_data['Age'] = titanic_data['Age'].fillna(age_med)
test_data['Age'] = test_data['Age'].fillna(test_age_med)
```

Replaces missing Fare value with median

```python
fare_avg = test_data['Fare'].median() # 14.4542
test_data['Fare'] = test_data['Fare'].fillna(fare_avg)
```

- Creates list of features to make predictions off of
- Sets up X and y to used in model
- Sets up test_X to be used in predictions

```python
features = ['Pclass', 'Sex', 'Age', 'Fare', 'Embarked']
X = titanic_data[features]
y = titanic_data['Survived']
test_X = test_data[features]
```

Confirming that all nulls have been filled

```python
X.info()
test_X.info()
```

<a id='rf'></a>
### First model: RandomForestRegressor
- Creates an empty RandomForestRegressor to cross-validates data by dividing it and checking 5 different subsets of the data set
- Creates and fits RandomForestRegressor model, then makes predictions by comparing final to test_X

```python
# produced best score of all models using only Sex and Pclass
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(random_state = 1)
cv = cross_val_score(rf, X, y, cv = 5)
print(cv.mean()) # average model score
final = rf.fit(X, y)
predictions = final.predict(test_X).round().astype(int)
# rounding/casting to integers to have correct form for answers for the competition
predictions
```

<a id='xgb'></a>
### Second model: XGBoost Regressor
- Creates an empty XGBRegressor to cross-validates data by dividing it and checking 5 different subsets of the data set
- Creates and fits XGBRegressor model, then makes predictions by comparing final to test_X

```python
from xgboost import XGBRegressor
xgb = XGBRegressor(random_state = 1)
cv = cross_val_score(xgb, X, y, cv = 5)
print(cv.mean())
final = xgb.fit(X, y)
predictions = final.predict(test_X).round().astype(int)
predictions
```

<a id='xgbc'></a>
### Third model: XGBoost Classifier
- Creates an empty XGBClassifier to cross-validates data by dividing it and checking 5 different subsets of the data set
- Creates and fits XGBClassifier model, then makes predictions by comparing final to test_X
- Better model type because this is a classification problem, not a regression problem

```python
# there's an error every time XGBClassifier runs but it still works
from xgboost import XGBClassifier
xgbc = XGBClassifier(random_state = 1, use_label_encoder = False)
cv = cross_val_score(xgbc, X, y, cv = 5)
print(cv.mean())
final = xgbc.fit(X, y)
predictions = final.predict(test_X)
predictions
```

<a id='rfc'></a>
### Fourth model: RandomForestClassifier
-  Creates an empty RandomForestClassifier to cross-validates data by dividing it and checking 5 different subsets of the data set
- Creates and fits RandomForestClassifier model, then makes predictions by comparing final to test_X

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(random_state = 1)
cv = cross_val_score(rf, X, y, cv = 5)
print(cv.mean())
final = rfc.fit(X, y)
predictions = final.predict(test_X)
predictions
```
