# Project 2: College Majors

### Downloading the data and creating the data frame

```python
import pandas as pd
rgd = pd.read_csv('recent-grads.csv', index_col = 0)
df = pd.DataFrame(rgd)
pd.set_option('display.max_columns', None)
df
```

## Cleaning the data

```python
# https://datatofish.com/rows-with-nan-pandas-dataframe/
df[df.isnull().any(axis=1)]
df = df.dropna()
df
```

Removes any null values to be able to more accurately interpret the data.

### Describing the data

```python
df.describe()
```

Some initial observations: 
- On average, the number of women in a major is higher than the number of men.
- The average ShareWomen is 0.52, but the standard deviation is 0.23, so there is still a lot of division between men and women in majors.

```python
df.corr()
```

Observations:
- Pretty high correlation between ShareWomen and Rank - women are working lower-paid jobs
- Almost no correlation between the number of men in a major and the ranking - men are working a variety of jobs at different pay grades

### What are women (not) majoring in?

```python
df.nlargest(10, 'ShareWomen')
```

Finds the 10 majors with the highest ShareWomen, which are predominantly in health, education, and psychology & social work. 

```python
df.nsmallest(10, 'ShareWomen')
```

Finds the 10 majors with the lowest ShareWomen, which are predominantly in engineering

```python
highestShrWmn = len(df.loc[(df.ShareWomen > 0.5), ['Major', 'ShareWomen']])
highestShrWmn
```

Shows that more than half of the majors are predominantly women, which is consistent with earlier observations

## Creating a series

```python
majCats = pd.Series(df['Major_category']).unique()
majCats
```

Creates a series which contains all of the different categories of majors in the data frame

### What is the average ShareWomen in each Major_category?

```python
means = []

for category in majCats: 
    maj = df.loc[df.Major_category == category]
    majMean = sum(maj.ShareWomen) / len(maj)
    means.append(majMean)

meanShrWmn = pd.Series(means, index = majCats)
meanShrWmn
```

Creates an empty array to store the average ShareWomen for each major, then loops through all of the categories to find their averages and add them to the array. Then creates a series using the averages and uses the series containing all the Major_category data as the index. 

```python
meanShrWmn.sort_values(ascending=False)
```

Sorts the average ShareWomen in each category from high to low. 

The major categories with the highest ShareWomen are health, psychology & social work, interdisciplinary (only 1 major in this category), and education, which is consistent with the categories of the 10 most popular majors among women. 

The major category with the lowest ShareWomen is engineering, which is consistent with the 10 least popular majors among women. 

This goes back to the high standard deviation. Even though the average ShareWomen overall is about half, some types of majors are significantly more popular among women. 

## What kinds of majors pay the most? 

```python
avgMedians = []

for category in majCats: 
    maj = df.loc[df.Major_category == category]
    majMean = sum(maj.Median) / len(maj)
    avgMedians.append(majMean)

meanMedian = pd.Series(avgMedians, index = majCats)
meanMedian
```

Uses the same method from finding the average ShareWomen in each Major_category to find the average Median in each Major_category. 

```python
meanMedian.sort_values(ascending=False)
```

Ranks the each Major_category from highest to lowest average Median. 

Engineering (the Major_category with the lowest ShareWomen) pays the most. 

Psychology & Social Work (the Major_category with the second-highest ShareWomen) pays the least. 

```python
pd.concat([meanShrWmn, meanMedian], axis=1)
```

Appends the two series to be compared and ranks them based on average median salary. 

The majors with the highest ShareWomen all rank below the majors with a lower ShareWomen. This is consistent with the correlation between ShareWomen and lowering rankings in the original data.
