```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
people = pd.read_csv('people.csv')
people
```

```python
sns.lmplot(x = 'Friends', y = 'GPA', data = people)
plt.show()
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv', index_col = 0)
reviews
```

```python
mostPoints = reviews.loc[(reviews.points == 100)]
bardata = mostPoints.value_counts('country')

sns.barplot(x = bardata.index, y = bardata)
plt.title('Number of best wines by country', fontsize = 20)
plt.grid(axis= 'y')
plt.show()
```
